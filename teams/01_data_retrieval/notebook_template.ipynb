## Step 1 - Verify SSD Access and Dataset Paths

```python
import os
from pathlib import Path

# Edit this to match your SSD mount point
SSD_ROOT = Path("/mnt/ssd")  # Linux, or change to /Volumes/SSD on Mac

DATASET_DIRS = {
    "thaillm":     SSD_ROOT / "data" / "thaillm",
    "seapile":     SSD_ROOT / "data" / "seapile",
    "thestack_v2": SSD_ROOT / "data" / "thestack_v2",
    "finemath":    SSD_ROOT / "data" / "finemath",
}

for name, path in DATASET_DIRS.items():
    status = "OK" if path.exists() else "NOT FOUND"
    print(f"{name:20s}: {status} -> {path}")
```

## Step 2 - Count Files and Estimate Total Size

```python
import subprocess

ASSIGNED_DATASET = "thaillm"  # Change to your assigned dataset

dataset_path = DATASET_DIRS[ASSIGNED_DATASET]
files = sorted(dataset_path.glob("*.jsonl.gz"))
total_bytes = sum(f.stat().st_size for f in files)

print(f"Dataset     : {ASSIGNED_DATASET}")
print(f"Shard count : {len(files)}")
print(f"Total size  : {total_bytes / 1e9:.2f} GB")
print()
for f in files[:5]:
    print(f"  {f.name:40s}  {f.stat().st_size / 1e6:.1f} MB")
if len(files) > 5:
    print(f"  ... and {len(files) - 5} more")
```

## Step 3 - Quick Schema Peek (First 3 Docs of First Shard)

```python
import gzip, json

first_shard = files[0]
print(f"Peeking at: {first_shard.name}")
print("-" * 60)

with gzip.open(first_shard, "rt", encoding="utf-8") as f:
    for i, line in enumerate(f):
        doc = json.loads(line)
        print(f"Doc {i}: keys = {list(doc.keys())}")
        print(f"  text[:120] = {doc.get('text', '')[:120]}")
        print()
        if i >= 2:
            break
```
